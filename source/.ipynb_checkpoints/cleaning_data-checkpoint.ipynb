```python
# import libraries 
import numpy as np 
import pandas as pd 
```

```python
# read data from excel file
data_2015 = pd.read_excel('../datos/2015.xlsx', header=None)
data_2015.drop([1,2,3], axis=1, inplace=True)
# show the first 5 elements of the dataset
data_2015.head()
```

```python
col_names = {0: 'estado', 4: 'sem', 5: 'm_acum', 6: 'f_acum', 7: 'acum'}
data_2015.rename(columns=col_names, inplace=True)
```
