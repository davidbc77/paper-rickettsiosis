```python
data2016 = '../Datos/2016.xlsx'

# retrieve data from all the sheets
xls = pd.ExcelFile(data2016)

semanas = xls.sheet_names

listings = []    # create empty list to collect DataFrames

col_names = {0: 'estado', 4: 'sem', 5: 'm_acum', 6: 'f_acum', 7: 'acum'}

for semana in semanas:
    listing = pd.read_excel(xls, sheet_name=semana, header=None)    # read a sheet
    listing.drop([1,2,3, 8], axis=1, inplace=True)    # delete colums unnecesary columns
    listing['semana'] = semana.split('sem')[-1]    # add the 'semana' column
    listing['year'] = 2016
    listing.rename(columns=col_names, inplace=True)    # rename columns
    listing = listing[:].replace('-',0)    # replace '-' for 0
    listings.append(listing)    # appends all DataFrames
```

```python
combined_listings = pd.concat(listings)
```

```python
combined_listings
```

```python
combined_listings.to_csv('2016.csv', index=False)
```

```python
combined_listings.describe()
```
